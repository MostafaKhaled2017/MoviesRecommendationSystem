```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
columns_u_data = ['user_id', 'item_id', 'rating', 'timestamp']
u_data = pd.read_csv('../data/raw/ml-100k/u1.base', sep='\t', names=columns_u_data)

columns_u_user = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
u_user = pd.read_csv('../data/raw/ml-100k/u.user', sep='|', names=columns_u_user)

columns_u_item = ['movie_id', 'title', 'release_date', 'video_release_date', 'IMDB_URL',
                  'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy',
                  'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
                  'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
u_item = pd.read_csv('../data/raw/ml-100k/u.item', sep='|', names=columns_u_item, encoding='latin-1')
```

## User Demographics Visualizations

```python
figures_path = '../reports/figures/'

# Age Distribution of Users
plt.figure(figsize=(10, 6))
sns.histplot(x='age', data=u_user, bins=30, kde=True)
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Count')

# Save the visualization
plt.savefig(figures_path + 'age_distribution.png')

# Show the Visualization
plt.show()
```

```python
# Gender Distribution of Users
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=u_user)
plt.title('Gender Distribution of Users')
plt.xlabel('Gender')
plt.ylabel('Count')

# Save the visualization
plt.savefig(figures_path + 'gender_distribution.png')

# Show the Visualization
plt.show()
```

## Movie Ratings and Genres

```python
# Distribution of Movie Ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=u_data)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')

# Save the visualization
plt.savefig(figures_path + 'movies_ratings_distribution.png')

# Show the Visualization
plt.show()
```

```python
# Distribution of Movie Genres
genre_columns = ['Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary',
                 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
                 'Thriller', 'War', 'Western']

genre_counts = u_item[genre_columns].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=genre_counts.values, y=genre_counts.index, orient='h')
plt.title('Distribution of Movie Genres')
plt.xlabel('Count')
plt.ylabel('Genre')

# Save the visualization
plt.savefig(figures_path + 'movies_genres_distribution.png')

# Show the Visualization
plt.show()
```

##  Movie Ratings Analysis

```python
# Distribution of Average Movie Ratings
average_ratings = u_data.groupby('item_id')['rating'].mean()
plt.figure(figsize=(10, 6))
sns.histplot(average_ratings, bins=50, kde=True)
plt.title('Distribution of Average Movie Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Count')

# Save the visualization
plt.savefig(figures_path + 'average_movie_ratings_distribution.png')

# Show the Visualization
plt.show()
```

```python
# Distribution of Ratings per Movie
ratings_per_movie = u_data.groupby('item_id')['rating'].count()
plt.figure(figsize=(10, 6))
sns.histplot(ratings_per_movie, bins=50, kde=True)
plt.title('Distribution of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Count')

# Save the visualization
plt.savefig(figures_path + 'ratings_per_movie_distribution.png')

# Show the Visualization
plt.show()
```
